## Linear Search

```python
target = 101
lst = [100,20,30,40,50,10]

found = False

for i in range(len(lst)):
    if lst[i] == target:
        print(f"Found {target} at index {i}")
        found = True
        break


if not found:    
    print(f"Target not found in the list")
        
```

```python
def linear_search(arr, target):
    size = len(arr)
    for index in range(0,size):
        if(arr[index] == target):
            return index

    return -1

arr = [10,20,30,40,50]
target = 40

result = linear_search(arr, target)
print(result)
```
